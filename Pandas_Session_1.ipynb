**Inplace**

-  When we do the operation the result output is not overwritting.

-  the original dataframe is same as it is.

-  in order to do this we need to use **inplace**.

-  by default inplace = False.

-  make inplace = True if you want to overwrite the result.

```python
df2.drop('Gender',axis=1,inplace=True)
```

```python
df2
```
