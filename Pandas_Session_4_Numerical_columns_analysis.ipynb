**Winsorization**

- whatever the ouliers less than the lower bound filled with lower bound
  
- whatever the ouliers less than the upper bound filled with upper bound

- In numpy we have the Method **clip**

```python
pwage_data=visa_df['prevailing_wage']
values=visa_df['prevailing_wage'].values.tolist()
p25=np.percentile(pwage_data,25)
p50=np.percentile(pwage_data,50)
p75=np.percentile(pwage_data,75)
iqr=p75-p25
lb=p25-1.5*iqr
ub=p75+1.5*iqr
new_Pwage_data=[np.clip(pwage_data,lb,ub)]
new_Pwage_data
```

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1).hist(pwage_data)
plt.subplot(2,2,2).hist(new_Pwage_data)
plt.subplot(2,2,3).boxplot(pwage_data,vert=False)
plt.subplot(2,2,4).boxplot(new_Pwage_data,vert=False)
plt.show()
```

```python
pwage_data=visa_df['prevailing_wage']
values=pwage_data.values.tolist()
p25=np.percentile(pwage_data,25)
p50=np.percentile(pwage_data,50)
p75=np.percentile(pwage_data,75)
iqr=p75-p25
lb=p25 - 1.5 * iqr
ub=p75 + 1.5 * iqr
con1=pwage_data>ub
con2=pwage_data<lb
con=con1 | con2
data=visa_df[con]
d=data['prevailing_wage'].values.tolist()
new_pwage=[]
for i in pwage_data:
    if i in d:
        if i < lb:
            new_pwage.append(lb)
        elif i > ub:
            new_pwage.append(ub)
        else:
            new_pwage.append(i)
new_pwage
```

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1).hist(pwage_data)
plt.subplot(2,2,2).hist(new_pwage)
plt.subplot(2,2,3).boxplot(pwage_data,vert=False)
plt.subplot(2,2,4).boxplot(new_pwage,vert=False)
plt.show()
```
