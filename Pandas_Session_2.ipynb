```python
dtype=visa_df.dtypes.values.tolist()
cols=visa_df.dtypes.keys().tolist()
cats=[]
for i,j in zip(cols,dtype):
    if j == 'object':
        cats.append(i)
print(cats)
```

```python
cats1=[j for i,j in zip(dtype,cols) if i =='object']
cats1
```

```python
num=[j for i,j in zip(dtype,cols) if i !='object']
num
```

```python
len(cats),len(cats1),len(num)
```
